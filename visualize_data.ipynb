```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data/20231228_sensor_data.csv")
```

```python
df.head()
```

Convert timestamp column 't' to datetime

```python
df['t'] = pd.to_datetime(df['t']/1000., unit='s')
```

```python
df.head()
```

# Basic visualisations

```python
import plotly.express as px

df_online = df[df['online']]

fig = px.scatter(
    df_online,
    x='t',
    y='cur_power'
)
fig.show()

fig = px.scatter(
    df_online,
    x='t',
    y='add_ele'
)
fig.show()
```
